```python
import pandas as pd
import numpy as np

import math
import time
import gc
import pickle
from numpy import array
```

```python
with open('sub_XGB_old', 'rb') as f:
    y_predsall2 = pickle.load(f)
    print(y_predsall2[0])
```

```python
out = 'submission1130.csv'
```

```python
df_test_dtypes = {'building_id': np.uint16, 'meter': np.uint8}
test = pd.read_csv('files/test.csv',dtype = df_test_dtypes)
test2 = test['meter'].copy()
df_building_metadata_dtypes = {'site_id': np.uint8, 'building_id': np.uint16, 'square_feet': np.int32, 'year_built': np.float32, 'floor_count': np.float32}
info = pd.read_csv('files/building_metadata.csv', dtype = df_building_metadata_dtypes)
weather_test = pd.read_csv('files/weather_test4.csv')
```

```python
test2 = test[['row_id','meter']].copy()
```

```python
# create list of building_id and meter
meter_pair = test[['building_id','meter']]
meter_pair = meter_pair.groupby(['building_id', 'meter']).size().reset_index(name='Freq')
meter_pair = meter_pair[['building_id', 'meter']].values.tolist()
print(len(meter_pair))
# check if 2380
```

```python
tick = time.time()
cnt = 0
last_id = -1
for (building_id,meter) in meter_pair:
#     if building_id < 100:
#         cnt = cnt+1
#         continue
    print(building_id,meter)
    if building_id != last_id:
        last_id = building_id
        info2 = info[info['building_id'] == building_id]
        site_id = int(info2['site_id'])
        timestamp = weather_test[weather_test['site_id'] == site_id]['timestamp']
    temp = (test['building_id']==building_id) & (test['meter']==meter)
#     diff = len(timestamp)-len(y_predsall2[cnt])
    a = test[temp]
    temp2 = a['timestamp'].isin(timestamp)
    a.loc[temp2,'meter'] = y_predsall2[cnt]
    if(temp2.sum() != 17520):
        fill1 = a['meter'].fillna(method='ffill',downcast='float32')
        fill2 = a['meter'].fillna(method='bfill',downcast='float32')
        fill1=fill1[~temp2]
        fill2=fill2[~temp2]
        a.loc[~temp2,'meter'] = (fill1 + fill2)/2
    test2.loc[temp,'meter'] = a['meter']
    cnt = cnt + 1
tock = time.time()
print('Time used {}.'.format(tock-tick))
```

```python
test2 = test2[['row_id','meter']]
test2.columns = ['row_id', 'meter_reading']
# 4dp required
test2['meter_reading'] = np.round_(test2['meter_reading'], decimals=4, out=None)
```

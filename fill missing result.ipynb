```python
import pandas as pd
import numpy as np

import math
import time
import gc
import pickle
from numpy import array
```

```python
# sub = xgboost
# sub2 = lgbm
with open('sub2', 'rb') as f:
    submission = pickle.load(f)
out = submission1214_1.csv
```

```python
# create list of building_id and meter
meter_pair = submission[['building_id','meter']]
meter_pair = meter_pair.groupby(['building_id', 'meter']).size().reset_index(name='Freq')
meter_pair = meter_pair[['building_id', 'meter']].values.tolist()
print(len(meter_pair))
# check if 2380
```

```python
submission['predicted'].sum()
```

```python
tick = time.time()
tofill_all = np.array([])
fill_all = np.array([])

for (building_id, meter) in meter_pair:
    s = submission[submission['building_id'] == building_id]
    s = s[s['meter'] == meter]
    if s['predicted'].sum() == 17520:
        continue
        
    print(building_id, meter)
    s2 = s.copy()
    s2 = s2.sort_values(by=['timestamp'])
    tofill = (s2['predicted'] == 0)
    s2.loc[tofill,'meter_reading'] = np.nan
    fill1 = s2['meter_reading'].fillna(method='ffill',downcast='float32')
    fill2 = s2['meter_reading'].fillna(method='bfill',downcast='float32')
    fill1 = fill1[tofill]
    fill2 = fill2[tofill]
    fill1 = fill1.fillna(method='bfill',downcast='float32')
    fill2 = fill2.fillna(method='ffill',downcast='float32')
    fill1 = (fill1+fill2)/2
    tofill = tofill[tofill]
    tofill = tofill.index
    tofill_all = np.concatenate((tofill_all, tofill), axis=None)
    fill_all = np.concatenate((fill_all, fill1), axis=None)
tock = time.time()
print('Time used : {}'.format(tock-tick))
```

```python
tick = time.time()
submission.loc[tofill_all,'meter_reading'] = fill_all
submission.loc[tofill_all,'predicted'] = 1
tock = time.time()
print(tock-tick)
```

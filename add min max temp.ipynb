```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import time
```

```python
# in_csv = 'files/weather_train_fixweather.csv'
# out_csv = 'files/weather_train_agg.csv'
in_csv = 'files/weather_test_fixweather.csv'
out_csv = 'files/weather_test_agg.csv'

weather = pd.read_csv(in_csv)
weather2 = weather.copy()
```

```python
display(weather2)
```

```python
weather2['timestamp'] = pd.to_datetime(weather2['timestamp'])
weather2['day'] = pd.to_datetime(weather2['timestamp']).apply(lambda x: x.date())
```

```python
ref = weather2.loc[0]['day']
weather2['day'] = weather2['day'] - ref
weather2['day'] = weather2['day'].dt.days
display(weather2)
print(weather2.dtypes)
```

```python
weather3 = weather2.copy()
```

```python
temp_stat = weather2.groupby(['day','site_id'])['air_temperature'].describe()
```

```python
display(temp_stat)
```

```python
t = temp_stat['min']
t = t.rename("min_temperature") 
weather3 = weather3.merge(t,on = ['day','site_id'] ,how = 'left')
t = temp_stat['max']
t = t.rename("max_temperature") 
weather3 = weather3.merge(t,on = ['day','site_id'],how = 'left')
t = temp_stat['mean']
t = t.rename("mean_temperature") 
weather3 = weather3.merge(t,on = ['day','site_id'],how = 'left')
```

```python
display(weather3)
```

```python
weather3.to_csv(out_csv, index=False)
```

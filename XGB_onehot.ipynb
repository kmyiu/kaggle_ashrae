```python
with open('train_1124', 'rb') as f:
    OH_X_train = pickle.load(f)
with open('test_1124', 'rb') as f:
    OH_X_test = pickle.load(f)
```

```python
OH_X_train = OH_X_train.drop(['has_year','has_floorcnt'],axis=1)
OH_X_test = OH_X_test.drop(['has_year','has_floorcnt'],axis=1)
```

```python
with pd.option_context('display.max_rows', None):
    print(OH_X_train.dtypes)
```

```python
OH_X_train['hour'] = pd.Series(OH_X_train['hour']/3, dtype='uint8')
OH_X_test['hour'] = pd.Series(OH_X_test['hour']/3, dtype='uint8')
```

```python
onehot_list = ['primary_use','month','weekday','hour','wind_direction']
```

```python
OH_X_train = pd.get_dummies(OH_X_train,columns=onehot_list)
OH_X_test = pd.get_dummies(OH_X_test,columns=onehot_list)
```
